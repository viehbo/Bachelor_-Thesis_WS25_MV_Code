# Glaciers as a scale of climate change





#### Preamble

One of the most crucial and challenging parts in AI is data selection. We are often confronted with an huge amount of data, where they are not specifically made for an machine learning task.
In this case weather data are to be observed. The variety and the amount of available data is far beyond every capability of non-supercomputer sized systems. Therefore, good data selection is necessary before starting any konf of model selection.

#### Objective 

For a potential machine learning task, only necessary weather data have to be selected.
We want to have a tool to visualize different kinds of weather data, like temperature, wind, humidity. This is to observe, if they can contribute information or not.

Good data selection saves time, money and our environment.
