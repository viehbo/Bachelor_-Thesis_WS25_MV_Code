# Scope of the project 

This document lists the whole scope of the practical work and the bachelor thesis in AI. It will be maintained as long the project/thesis is running. Therefore, it's the core resource about the content.



## 1. Project Description and Overview
The idea is to provide a user interface, where one can compare different meteorological with the change of glaciers. It can be seen as a data selection task.

The whole project can be separated into 3 main parts. As there are:

* ##### Data collection
  For the weather data the ERA5 dataset is used. So they are already collected. The data about the glaciers have to be collected from different sources.

* ##### Data processing
  Of both, weather data and the glacier data

* ##### User Interface
  Provide an intuitive user interface. The data should be represented in tabular form to sort them and in a graphical way on a map.


## 2. Roadmap for "Practical Work in AI"
Here is a brief procedure of the project. Detailed informations are described in the corresponding packages.

#### 1 Restriction of the data and the area
The ERA5 dataset has a resolution of 1-3h on a 9km grid with 50 features. Most likely, the ERA5 Global Atmosphere data are also needed for this project. In Order to this huge amount of data, only 5 features should be used for the visualization.

The Alps have still about 5000 glaciers. Out of this, 5 regions with 3-8 glaciers each have to be chosen. The regions should be spread over the Alps.


#### 2 Get the weather data
To get started, the chosen features should be downloaded in a low resolution for a short time period. This should speed up the development of the visualization. Later, an API request function have to be implemented. If as an addition or an optional feature, have to be decided later.


#### 3 Collect the glacier data
Goal is, to find 5 regions with glaciers, where representative data exist. This will take some time on research.
The glaciers in each region should have different alignments and different heights. So the regions can be comparable.


#### 4 Visualize the data
Now it's time to create a map. The chosen glaciers should be obviously highlighted. But also the whole mountain area of the Alps have to be covered in order to locate them.
At this point, the way to represent the data have to be determent. Multiple options and there combinations are possible, e.g. color, circles or arrows and there size, ...

The weather data are on a second overlapping layer. Color and hue might be a good starting point.

As mentioned before, the data should also be represented in tabular form.

This task is directly connected to the next one.


#### 5 User interaction
The following functions must be implemented:

* shifting and panning the map (tilting should not be possible)
* hover the pointer over a glacier provides the name and some basic information
* clicking on a glacier shows detailed information
* the user can adjust the observed time frame
* in this time frame, the features can be selected just for night day time


#### 6 API integration
Up to this point, only local stored data are used. To make this tool versatile, requests over the ERA API must be implemented.
The user should only be able to request a certain amount of data. Otherwise, the request time get's too long or causing overflow problems.
